### Rellenar datos faltantes - Imputar

Imputar con:
- **Media**: Promedio de los datos (se ve afectada por valores extremo)
- **Mediana:** Valor que se encuentra justamente en la posición central de la serie de datos ordenada
- **Moda:** El mas frecuente (puede ser unimodal, bimodal, multimodal o no haber moda)

```python
datosRelleno = copiaDatos.copy()
datosRelleno['age'] = datosRelleno['age'].fillna(datosRelleno['age'].mean())
datosRelleno.info()
```

```python
# Segmentar los grupos en entrenamiento y pruebas
y = datosRelleno['survived']
datosRelleno.drop('survived', axis=1, inplace=True)
x_train, x_test, y_train, y_test = train_test_split(datosRelleno, y, test_size=0.3, random_state=50)
lr.fit(x_train, y_train)
pred = lr.predict(x_test)
print(metrics.accuracy_score(pred, y_test));
```

### Detección de valores atipicos utilizando el rango intercuartil (RIC)

$RIC = Q_3 - Q_1$

$L_i = Q_1 - 1.5 X RIC$

$L_s = Q_3 + 1.5 X RIC$

```python
outliers = []
```

```python
def detectar_outliers_ric(datos):
    datos = sorted(datos)
    q1 = np.quantile(datos, 0.25)
    q3 = np.quantile(datos, 0.75)
    ric = q3 - q1
    li = q1 - 1.5*ric
    ls = q3 + 1.5*ric
    print(f'Limite inferior: {li}; limite superior: {ls}')
    for valor in datos:
        if (valor < li or valor > ls):
            outliers.append(valor)
    return outliers
```
