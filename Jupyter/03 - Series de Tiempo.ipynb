```python
import yfinance as yf

# Definir el ticker
ticker = "GOOGL"

# Descargar histÃ³rico (ejemplo: desde 2020 hasta hoy)
df_google_2025 = yf.download(ticker, start="2006-01-01", end="2016-09-26", interval="1d")

# Imprimir el DataFrame completo
display(df_google_2025);
```

```python
df_google_2025['Close'].plot()
plt.title(f"Precio de cierre de {ticker}")
plt.xlabel("AÃ±o")
plt.ylabel("Precio de cierre")
plt.show()
```

```python
df_google_to_csv = df_google_2025.to_csv('google_2025.csv')
```
